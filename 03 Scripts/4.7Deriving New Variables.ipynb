<H3>If Statement with user defined function</H3>

```python
#Define a function

def price_label(row):

    if row ['prices']<= 5:
        return 'Low range product'
    elif (row['prices']> 5) and (row['prices']<=15):
        return 'Mid range product'
    elif row['prices'] >15 :
        return 'High range product'
    else:return 'Not enough data'
```

```python
#Apply the function

df['price_range'] =df.apply(price_label,axis =1)
```

```python
#Check output

df['price_range'].value_counts(dropna = False)
```

<H4>Update "busiest day" column to "busiest days" to identify order day of the week as \"Busiest days,\" \"Slowest days,\" or \"Regularly busy</H4>

```python
# Create column 'busiest_days" using the frequency of "orders_days_of_week" to identify two busiest days (0, 1), slowest days (4, 3), or regularly busy (all other days)

result_2 =[]

for value in ords_prods_merge["orders_days_of_week"]:
    if value ==0 or value==1:
        result_2.append("Busiest Days")
    elif value==4 or value==3:
        result_2.append("Slowest Days")
    else:
        result_2.append("Regularly busy")
```

<H4>When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, periods of time to be labeled Most orders, Average orders, and Fewest orders.Create a new column containing these labels called busiest_period_of_day.</H4>

```python
#Check frequency of the 'order_hour_of_day' column

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create a new column 'Busiest_period_of_day' with the following three values
# * Most orders (9 - 16) 
# * Average orders (7-8 and 17 - 22)
# * Fewest orders  (23,0- 6)


result3 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [9,10,11,12,13,14,15,16]:
      result3.append("Most orders")
  elif value in [17,7,8,18,19,20,21,22]:
      result3.append("Average orders")
  else:result3.append("Fewest orders")
```

```python
result3
```

```python
#Create new column busiest_period_of_day using result3
ords_prods_merge['busiest_period_of_day'] = result3
```
