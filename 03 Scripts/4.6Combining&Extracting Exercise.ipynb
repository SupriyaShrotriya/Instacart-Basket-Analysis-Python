<H3>Merging the DataFrames</H3>

```python
#Clean the dataframes before merging by dropping the unnecessary columns 

df_orders_products_combined = df_orders_products_combined.drop(['Unnamed: 0', '_merge'], axis =1)
```

```python
df_orders_products_combined
```

```python
df_prods = df_prods.drop(['Unnamed: 0'], axis =1)
```

```python
df_prods
```

<H4>Observation -The 'product_id' column is present in both datasets allowing us to merge both the datasets together.We information that exists in both dataframes, therefore, the 'inner' merge will be a suitable fit. </H4>

```python
# Test merge without overwriting with merge flag 'match'

pd.merge(df_orders_products_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match' )
```

<H5>The test merge indicates that the merge with inner join still gives all the 32404859 rows so goign ahead with the actual merge </H5>

```python
# Actual merge of dataframes 'df_orders_products_combined' and 'df_prods'

ords_prods_merge = df_orders_products_combined.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```
